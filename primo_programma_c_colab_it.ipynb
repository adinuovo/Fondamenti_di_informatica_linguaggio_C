# Primo programma in C in Google Colab

Questo notebook spiega un semplice programma in C e ti permette di compilarlo ed eseguirlo direttamente in Colab.


## Il codice sorgente in C

```c
#include <stdio.h>   // Importa la libreria standard di input/output

int main() {         // Punto di ingresso del programma: l'esecuzione inizia qui
    printf("Ciao Mondo\n");  // Stampa il testo sul terminale
    return 0;        // Termina il programma indicando che l'esecuzione è avvenuta con successo
}
```


## Spiegazione

- `#include <stdio.h>` importa la libreria standard di input/output, necessaria per usare funzioni come `printf`.
- `main` è il punto di ingresso del programma: l’esecuzione inizia sempre da qui.
- `printf` stampa del testo sul terminale.
- `return 0` termina la funzione `main` e restituisce un codice di uscita al sistema operativo.

### Significato di `return 0`

In C, l’istruzione **`return 0;`** indica che il programma è terminato **correttamente**, senza errori.

- Il valore restituito da `main` viene passato al **sistema operativo**.
- Per convenzione (specialmente su sistemi Unix/Linux):
  - **`0`** significa successo
  - **un valore diverso da 0** indica un errore o una condizione anomala

Esempi:
- `return 0;` → il programma è terminato correttamente
- `return 1;` → il programma è terminato con un errore

Nelle versioni moderne del linguaggio C (C99 e successive), se `main` termina senza un’istruzione `return`, il compilatore **assume automaticamente `return 0;`**. Tuttavia, scriverlo esplicitamente è considerata una **buona pratica**, soprattutto a scopo didattico.


## Compilazione ed esecuzione

Qui usiamo `%%writefile` per creare il file `primo.c` in modo semplice, poi compiliamo ed eseguiamo con `gcc`.


````python
%%writefile primo.c
#include <stdio.h>   // Importa la libreria standard di input/output

int main() {         // Punto di ingresso del programma
    printf("Ciao Mondo\n");  // Stampa il testo sul terminale
    return 0;        // Esecuzione terminata con successo
}
````

````python
!gcc primo.c -o primo
!./primo
````
